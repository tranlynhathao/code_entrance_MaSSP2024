## Part 1 (question 2.3)

Using Riemann sum to caculate
$$
\int_{lower}^{upper} func(x) dx
$$

````python
import numpy as np

def g(x):
    return np.exp(x) / (1 + np.exp(x)) - 0.5

def riemann_sum(a, b, n):
    dx = (b - a) / n
    x = np.linspace(a, b, n)
    return np.sum(g(x)) * dx

a = float(input("Nhập giá trị cận dưới của tích phân: "))
b = float(input("Nhập giá trị cận trên của tích phân: "))
n = int(input("Nhập số lượng điểm chia: "))

result = riemann_sum(a, b, n)
print("Kết quả tổng Riemann:", result)
````

## Part 2 (question 2.4)

**Note:**
- Please choose `random_seed` equal your day of birth with the following format: `ddmmyyyy`. For example if your day of birth is `20/02/2022`, your `random_seed` will be `20022002`
- `apt_dataset.answers` is a randomly generated dataframe containing the work of 100,000 fictitious students on an APT test. With 120 columns numbered from 1 to 120 corresponding to each student's answer to each question.
- `apt_dataset.solution` is the answer to this fictional APT contest.

````python
apt_dataset = APT_dataset(n_student=100000, random_seed=20022002)
````

````python
apt_dataset.answers
````

````python
apt_dataset.solution
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

answers = apt_dataset.answers
solution = apt_dataset.solution

# Tạo copy
answers_with_name = answers.copy()

if 'Name' in answers.columns:
    answers_without_name = answers.drop('Name', axis=1)
else:
    answers_without_name = answers

n_students = answers_without_name.shape[0]
n_questions = answers_without_name.shape[1]

# Giả sử tỷ lệ trả lời đúng mỗi câu theo phân phối đồng nhất
np.random.seed(42)  # Đặt seed để đảm bảo tính tái tạo
p_correct = np.random.uniform(0.1, 0.9, n_questions)

# Công thức
x_difficulty = np.log(p_correct / (1 - p_correct))
points_per_question = 7 + (x_difficulty - np.min(x_difficulty)) / (np.max(x_difficulty) - np.min(x_difficulty)) * (13 - 7)

# Tạo DataFrame mới để tính toán
df = pd.DataFrame(answers_without_name, columns=[f'Q{i+1}' for i in range(n_questions)])

# Tính điểm tổng cho từng thí sinh
correct_answers = (answers_without_name == solution).astype(int)
student_points = correct_answers.dot(points_per_question)

# Thêm cột điểm vào DataFrame ban đầu có chứa tên
answers_with_name['point'] = student_points

# Lấy 3 thí sinh có điểm cao nhất
top_students = answers_with_name.nlargest(3, 'point')
print("Ba thí sinh có điểm cao nhất:")
print(top_students[['Name', 'point']])
````
